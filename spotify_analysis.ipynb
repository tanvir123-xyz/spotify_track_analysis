```python
# 1. IMPORTS

import pandas as pd
import matplotlib.pyplot as plt


# 2. LOAD DATASET

df = pd.read_csv('spotify_tracks.csv')

print("Shape of dataset:", df.shape)
print("Columns:", df.columns.tolist())


# 3. HEAD

print("First 5 rows:")
display(df.head())

# Shape of dataset
print("Dataset shape (rows, columns):")
print(df.shape)

# List of columns
print("Column names:")
print(df.columns.tolist())

# Data types of each column
print("\nData types:")
print(df.dtypes)

# Basic summary of numerical attributes
print("\nStatistical summary:")
display(df.describe())


```

```python
# List of numerical columns to summarize
cols = ["popularity","danceability","energy","acousticness","instrumentalness",
        "speechiness","valence","loudness","tempo","duration_ms"]

# Generate individual summaries
for c in cols:
    print(f"\nStatistical Summary for: {c}\n")
    print(df[c].describe())
```

```python
print("Before removing duplicates:", df.shape[0])

# Remove duplicate tracks based on track_id or track_name + artist_name
df_dedup = df.drop_duplicates(subset=["track_id"], keep="first")

# Number of tracks after removing duplicates
print("After removing duplicates:", df_dedup.shape[0])

# Number of removed duplicate records
print("Duplicates removed:", df.shape[0] - df_dedup.shape[0])
```

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(
    df['popularity'].dropna(),
    kde=True,
    color='#E74C3C',
    bins=25,
    edgecolor='black',
    linewidth=1.2
)
plt.title('Popularity Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Popularity', fontsize=12)
plt.ylabel('No. of songs', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df['danceability'].dropna(),
    kde=True,
    color='#00C49A',
    bins=25,
    edgecolor='black',
    linewidth=1.2
)
plt.title('Danceability Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Danceability', fontsize=12)
plt.ylabel('No. of Songs', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Categorize energy into bins
bins = [0, 0.33, 0.66, 1.0]
labels = ['Low Energy', 'Medium Energy', 'High Energy']
df['energy_level'] = pd.cut(df['energy'], bins=bins, labels=labels, include_lowest=True)

# Count percentage of each category
energy_counts = df['energy_level'].value_counts(normalize=True) * 100

# Plot pie chart
plt.figure(figsize=(6,6))
colors = ['#3498DB', '#F1C40F', '#E74C3C']   # blue, yellow, red
plt.pie(
    energy_counts,
    labels=energy_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 12}
)
plt.title('Energy Level Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['valence'], bins=25, color='skyblue',
             edgecolor='black', linewidth=1.2, alpha=0.7)
sns.kdeplot(df['valence'], color='blue', linewidth=2)

plt.title("Valence Distribution (Musical Positivity)", fontsize=14)
plt.xlabel("Valence (-ve = sad, +ve = happy)")
plt.ylabel("No. of songs")
plt.grid(axis='y', linestyle='--', alpha=0.35)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['tempo'], bins=30, color='lightgreen',
             edgecolor='black', linewidth=1.2, alpha=0.7)

#  average tempo
mean_tempo = df['tempo'].mean()
plt.axvline(mean_tempo, color='red', linestyle='--', linewidth=2,
            label=f"Mean = {mean_tempo:.1f}")

plt.title("Tempo Distribution (Beats Per Minute)", fontsize=14)
plt.xlabel("Tempo (BPM)")
plt.ylabel("No. of songs")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.35)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,4))

sns.violinplot(x=df['tempo'], color='lightgreen', linewidth=1.2)

# mean & median
mean_tempo = df['tempo'].mean()
median_tempo = df['tempo'].median()

# draw AFTER violin so it's on top
plt.axvline(mean_tempo, color='red', linestyle='--', linewidth=2, zorder=10, label=f"Mean = {mean_tempo:.1f}")
plt.axvline(median_tempo, color='blue', linestyle='-', linewidth=1, zorder=10, label=f"Median = {median_tempo:.1f}")

plt.title("Tempo Spread (BPM Distribution)", fontsize=14)
plt.xlabel("Tempo (BPM)")
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.35)
plt.tight_layout()
plt.show()
```

```python
# remove rows where language is missing / unknown
df = df[~df['language'].isin(["", "unknown", "Unknown", "UNK", "nan", "Not Available"])]
df = df.dropna(subset=['language'])
```

```python
import matplotlib.pyplot as plt

# Top 6 languages
lang_counts = df['language'].value_counts().head(5)

plt.figure(figsize=(6,6))
plt.pie(lang_counts, labels=lang_counts.index,
        autopct='%1.1f%%', startangle=140,
        wedgeprops={'linewidth': 1, 'edgecolor': 'black'})

# donut hole
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Top 5 Languages in Dataset")
plt.tight_layout()
plt.show()

```

```python
df['year_group'] = pd.cut(df['year'], bins=[1950,1980,2000,2010,2020,2025],
                          labels=['<1980','1980-2000','2000-2010','2010-2020','2020+'])
```

```python
year_counts = df['year_group'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(year_counts, labels=year_counts.index, autopct='%1.1f%%', startangle=140)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Song Distribution by Year Groups")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.regplot(
    x=df['energy'],
    y=df['popularity'],
    scatter_kws={'alpha':0.15, 'color':'orange', 's':12},
    line_kws={'color':'black', 'linewidth':2.5}
)

plt.title("Popularity vs Energy", fontsize=15, weight='bold')
plt.xlabel("Energy", fontsize=12)
plt.ylabel("Popularity", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
# create tempo categories
df['tempo_group'] = pd.cut(
    df['tempo'],
    bins=[0,80,110,140,200,300],
    labels=['Slow','Mid','Fast','Very Fast','Extreme']
)

plt.figure(figsize=(8,5))
sns.boxplot(x='tempo_group', y='popularity', data=df,
            palette='viridis', width=0.6)

plt.title("Popularity Across Tempo Categories", fontsize=15, weight='bold')
plt.xlabel("Tempo Category", fontsize=12)
plt.ylabel("Popularity", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(8,5))

sns.regplot(
    x=df['tempo'],
    y=df['popularity'],
    scatter_kws={'alpha':0.15, 'color':'purple', 's':12},
    line_kws={'color':'black', 'linewidth':2.5}
)

plt.title("Popularity vs Tempo (BPM)", fontsize=15, weight='bold')
plt.xlabel("Tempo (Beats Per Minute)", fontsize=12)
plt.ylabel("Popularity", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['duration_min'] = df['duration_ms'] / 60000


sns.regplot(
    x=df['duration_min'],
    y=df['popularity'],
    scatter_kws={'alpha':0.15, 'color':'brown', 's':12},
    line_kws={'color':'black', 'linewidth':2.5}
)

plt.title("Popularity vs Duration", fontsize=15, weight='bold')
plt.xlabel("Duration (min)", fontsize=12)
plt.ylabel("Popularity", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
df['energy_group'] = pd.cut(
    df['energy'],
    bins=[0,0.3,0.6,1.0],
    labels=['Low Energy','Medium Energy','High Energy']
)

df['dance_group'] = pd.cut(
    df['danceability'],
    bins=[0,0.3,0.6,1.0],
    labels=['Low Dance','Medium Dance','High Dance']
)

```

```python
plt.figure(figsize=(8,5))

pd.crosstab(df['energy_group'], df['dance_group']).plot(
    kind='bar',
    figsize=(8,5),
    edgecolor='black',
    colormap='viridis'
)

plt.title("Danceability vs Energy (Category Comparison)", fontsize=15, weight='bold')
plt.xlabel("Energy Category")
plt.ylabel("Number of Tracks")
plt.tight_layout()
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8,6))
corr = df[['popularity','danceability','energy','valence','tempo','duration_ms']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Selected Audio Features", fontsize=15, weight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['danceability'],
    df['popularity'],
    c=df['energy'],
    cmap='viridis',
    alpha=0.35,
    s=14
)

plt.colorbar(label='Energy')
plt.title("Popularity vs Danceability (Colored by Energy)", fontsize=15, weight='bold')
plt.xlabel("Danceability", fontsize=12)
plt.ylabel("Popularity", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(df['tempo'],df['energy'],df['danceability'],
    c=df['popularity'],cmap='viridis',alpha=0.4,s=12)

ax.set_title("3D: Tempo, Energy, Danceability (Color = Popularity)", fontsize=12)
ax.set_xlabel("Tempo (BPM)")
ax.set_ylabel("Energy")
ax.set_zlabel("Danceability")

cbar = plt.colorbar(sc)
cbar.set_label("Popularity")

plt.tight_layout()
plt.show()
```

```python
popular = df[df['popularity'] > 70]  # consider top popular tracks

year_features = popular.groupby('year')[['danceability', 'energy']].mean().reset_index()

plt.figure(figsize=(8,5))
plt.plot(year_features['year'], year_features['danceability'], marker='o', label='Danceability', color='green')
plt.plot(year_features['year'], year_features['energy'], marker='o', label='Energy', color='red')

plt.title("Danceability & Energy of Popular Songs Over Time", fontsize=15, weight='bold')
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
year_ac = df.groupby('year')[['acousticness','instrumentalness']].mean().reset_index()

plt.figure(figsize=(8,5))
plt.plot(year_ac['year'], year_ac['acousticness'], marker='o', label='Acousticness', color='blue')
plt.plot(year_ac['year'], year_ac['instrumentalness'], marker='o', label='Instrumentalness', color='orange')

plt.title("Acousticness & Instrumentalness Over Time", fontsize=15, weight='bold')
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# filter popular songs
popular = df[df['popularity'] > 70]

# compute average duration per year
year_popdur = popular.groupby('year')['duration_ms'].mean().reset_index()
year_popdur['duration_min'] = year_popdur['duration_ms'] / 60000

plt.figure(figsize=(8,5))
plt.plot(year_popdur['year'], year_popdur['duration_min'], marker='o', color='brown')

plt.title("Average Duration of Popular Songs Over Years", fontsize=15, weight='bold')
plt.xlabel("Year")
plt.ylabel("Average Duration (minutes)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
